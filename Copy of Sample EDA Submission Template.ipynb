#### Chart - 3

````python

# Set a style for the plots
sns.set_style('whitegrid')

# 1. Distribution of Credit_Score
plt.figure(figsize=(8, 5))
sns.countplot(x='Credit_Score', data=df, palette='viridis')
plt.title('Distribution of Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Count')
plt.show()

# 2. Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Distribution of Annual_Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual_Income'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()

# 4. Distribution of Credit_Mix
plt.figure(figsize=(8, 5))
sns.countplot(x='Credit_Mix', data=df, palette='viridis')
plt.title('Distribution of Credit Mix')
plt.xlabel('Credit Mix')
plt.ylabel('Count')
plt.show()

# 5. Distribution of Interest_Rate
plt.figure(figsize=(10, 6))
sns.histplot(df['Interest_Rate'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Interest Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Frequency')
plt.show()


````

#Let's visualize the relationships between some pairs of variables.



````python
# 4. Pie/Donut Chart – Department
#1 . Credit_Score vs Annual_Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Annual_Income', data=df, palette='viridis')
plt.title('Credit Score vs Annual Income')
plt.xlabel('Credit Score')
plt.ylabel('Annual Income')
plt.show()

# 2. Credit_Score vs Interest_Rate
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Interest_Rate', data=df, palette='viridis')
plt.title('Credit Score vs Interest Rate')
plt.xlabel('Credit Score')
plt.ylabel('Interest Rate')
plt.show()

# 3. Credit_Score vs Credit_Utilization_Ratio
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Credit_Utilization_Ratio', data=df, palette='viridis')
plt.title('Credit Score vs Credit Utilization Ratio')
plt.xlabel('Credit Score')
plt.ylabel('Credit Utilization Ratio')
plt.show()

# 4. Credit_Score vs Credit_History_Age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Credit_History_Age', data=df, palette='viridis')
plt.title('Credit Score vs Credit History Age')
plt.xlabel('Credit Score')
plt.ylabel('Credit History Age')
plt.show()

# 5. Credit_Score vs Num_of_Delayed_Payment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Num_of_Delayed_Payment', data=df, palette='viridis')
plt.title('Credit Score vs Number of Delayed Payments')
plt.xlabel('Credit Score')
plt.ylabel('Number of Delayed Payments')
plt.show()
````

Let's visualize the relationships between multiple variables.#### Chart - 5

````python
# Chart - 5 visualization code
# 1. Relationship between Annual_Income, Interest_Rate, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Annual_Income', y='Interest_Rate', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Annual Income vs Interest Rate colored by Credit Score')
plt.xlabel('Annual Income')
plt.ylabel('Interest Rate')
plt.show()

# 2. Relationship between Credit_Utilization_Ratio, Num_of_Delayed_Payment, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Credit_Utilization_Ratio', y='Num_of_Delayed_Payment', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Credit Utilization Ratio vs Number of Delayed Payments colored by Credit Score')
plt.xlabel('Credit Utilization Ratio')
plt.ylabel('Number of Delayed Payments')
plt.show()

# 3. Relationship between Age, Credit_History_Age, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='Credit_History_Age', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Age vs Credit History Age colored by Credit Score')
plt.xlabel('Age')
plt.ylabel('Credit History Age')
plt.show()

# 4. Relationship between Annual_Income, Credit_History_Age, and Credit_Score
plt.figure(figsize=(12, 8))
sns.boxplot(x='Credit_Score', y='Annual_Income', hue='Credit_Mix', data=df, palette='viridis')
plt.title('Credit Score vs Annual Income by Credit Mix')
plt.xlabel('Credit Score')
plt.ylabel('Annual Income')
plt.show()


# 5. Pairplot of selected numerical variables colored by Credit_Score
selected_cols = ['Annual_Income', 'Interest_Rate', 'Credit_Utilization_Ratio', 'Num_of_Delayed_Payment', 'Credit_Score']
sns.pairplot(df[selected_cols], hue='Credit_Score', palette='viridis')
plt.suptitle('Pairwise Relationships of Selected Variables by Credit Score', y=1.02)
plt.show()
````

#### Chart - 6

````python
# Chart - 6 visualization code
 # 6. Pie chart of Credit_Score distribution
credit_score_counts = df['Credit_Score'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(credit_score_counts, labels=credit_score_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(credit_score_counts)))
plt.title('Distribution of Credit Score')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

#### Chart - 7

````python
# Chart - 7 visualization code
# 7. Distribution of Monthly_Inhand_Salary
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly_Inhand_Salary'], bins=30, kde=True, color='purple')
plt.title('Distribution of Monthly Inhand Salary')
plt.xlabel('Monthly Inhand Salary')
plt.ylabel('Frequency')
plt.show()

# 8. Count of Payment_of_Min_Amount
plt.figure(figsize=(8, 5))
sns.countplot(x='Payment_of_Min_Amount', data=df, palette='viridis')
plt.title('Count of Payment of Minimum Amount')
plt.xlabel('Payment of Minimum Amount')
plt.ylabel('Count')
plt.show()

# 9. Relationship between Num_of_Loan and Credit_Score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Num_of_Loan', data=df, palette='viridis')
plt.title('Credit Score vs Number of Loans')
plt.xlabel('Credit Score')
plt.ylabel('Number of Loans')
plt.show()

# 10. Relationship between Outstanding_Debt and Credit_Score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Outstanding_Debt', data=df, palette='viridis')
plt.title('Credit Score vs Outstanding Debt')
plt.xlabel('Credit Score')
plt.ylabel('Outstanding Debt')
plt.show()

# 11. Count of Occupation
plt.figure(figsize=(12, 6))
sns.countplot(y='Occupation', data=df, order=df['Occupation'].value_counts().index, palette='viridis')
plt.title('Count of Occupation')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.show()
````
